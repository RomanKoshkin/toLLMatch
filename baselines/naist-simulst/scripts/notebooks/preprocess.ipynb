```python
from transformers import TrainingArguments, Trainer
import torch
import torch.nn as nn
from transformers import BertTokenizer, BertModel,PreTrainedModel
from datasets import load_dataset
import numpy as np
from datasets import load_metric,concatenate_datasets
from datasets.dataset_dict import DatasetDict
from datasets import Dataset
from transformers.modeling_outputs import ModelOutput
en_de=load_dataset('ahclab/acceptability_filtering_data_en-de',use_auth_token=True)['train']
en_zh=load_dataset('ahclab/acceptability_filtering_data_en-zh',use_auth_token=True)['train']
en_ja=load_dataset('ahclab/acceptability_filtering_data_en-ja',use_auth_token=True)['train']
en_de=en_de.rename_column("en_ids", "lang1").rename_column("de_ids", "lang2")
en_zh=en_zh.rename_column("en_ids", "lang1").rename_column("zh_ids", "lang2")
en_ja=en_ja.rename_column("en_ids", "lang1").rename_column("ja_ids", "lang2")
_data=concatenate_datasets([en_de,en_zh,en_ja])
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-uncased')
def connecter(x):
    src=x['lang1']
    tgt=x['lang2']
    if 0 in  src:
        src=x['lang1'][:x['lang1'].index(0)]
    
    if 0 in  tgt:
        tgt=x['lang2'][:x['lang2'].index(0)]
    ids=[tokenizer.cls_token_id] +src+[tokenizer.sep_token_id]+tgt
    ids_512=ids+[0]*(512-len(ids))
    return {'input_ids':ids_512,'label':x['label']}

data=_data.map(connecter)
data=data.remove_columns(['lang1','lang2','__index_level_0__'])
data=data.shuffle()
data.push_to_hub('ahclab/acceptability_filtering_data')
```
